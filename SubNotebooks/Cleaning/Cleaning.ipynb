### **| Fase 2 ) -** Geração de Prompts'

---

```python
import pandas as pd
import os

# 1. Definir e criar a pasta de destino
output_dir = 'files'
os.makedirs(output_dir, exist_ok=True)

# 2. Carregar o banco de dados (ajuste o caminho conforme o seu Jupyter Notebook)
df = pd.read_csv('../../Data/PosProcessed/gemma3_12b_3batch_v2.csv')

# Remover valores nulos na coluna de categorias para evitar falhas na iteração
df = df.dropna(subset=['Macro_Categoria'])

# Extrair a lista de todas as macro-categorias únicas da base
macro_categories = df['Macro_Categoria'].unique()

# 3. Template do prompt parametrizado
prompt_template = """Act as an expert in data science and behavioral modeling. Your task is to audit the classification of a subgroup of variables that were assigned to the macro-category: 
[{categoria}].

Below, I provide the list of variables extracted from my database that were classified under this label:
{variaveis}

Based on the strict rule provided, analyze the list and respond to the following points:

Rule Validation: Confirm if the meaning of each listed variable belongs exclusively to this category.

Anomaly Detection: Clearly list any variable that has been classified incorrectly or whose context is ambiguous.

Reallocation: For each incorrect variable, indicate which macro-category would be most appropriate and explain the structural reason.

Final Assessment: Give a brief evaluation regarding the purity of this data.

Final output: An table with the cols: oldvar, newvar
"""

# 4. Iterar sobre as categorias, preencher o texto e salvar em arquivos
for cat in tqdm(macro_categories):
    # Extrair a lista de variáveis pertencentes apenas à categoria atual
    var_list = df[df['Macro_Categoria'] == cat]['Col'].tolist()
    
    # Preencher o template com o nome da categoria e a lista gerada
    formatted_prompt = prompt_template.format(
        categoria=cat,
        variaveis=str(var_list)
    )
    
    # Tratar o nome do arquivo para não conter caracteres inválidos
    safe_filename = str(cat).replace(' ', '_').replace(',', '').replace('/', '_') + '.md'
    file_path = os.path.join(output_dir, safe_filename)
    
    # Escrever o conteúdo no arquivo Markdown
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(formatted_prompt)
```
